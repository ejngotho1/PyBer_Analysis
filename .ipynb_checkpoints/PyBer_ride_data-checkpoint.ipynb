```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create plot
plt.plot(x_axis, y_axis)
```

```python
# object oriented
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# create plot and add plt.show()
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# create plot and add plt.show()
plt.plot(x_axis, y_axis)
plt.plot(x_axis, y_axis, label='line')
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# create a bar graph
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot for a horizonatal bar
plt.barh(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot with jan on top y_axis
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()    #this inverts axis
```

```python
# Create the plot for a horizonatal bar
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
# invert axis values
plt.gca().invert_yaxis() 
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

#### add annotations to scatter

```python
# Create the plot for a scatter
plt.scatter(x_axis, y_axis, color="red", label='chicago')
# invert axis values
plt.gca().invert_yaxis() 
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot for a scatter and vary size of marker to the y-value (s=y_axis)
plt.scatter(x_axis, y_axis, s=y_axis, color="red", label='chicago')
# invert axis values
plt.gca().invert_yaxis() 
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot for a scatter and vary size of marker to the y-value (s=y_axis)
# then increase the size of marker by a parameter eg *....need to create a new axis and use it
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*10)
```

```python
## use this new axis for marker size
plt.scatter(x_axis, y_axis, s=y_axis_larger, color="red", label='chicago')
# invert axis values
plt.gca().invert_yaxis() 
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# alternatively use LIST COMPREHENSION as follows for size
## use this new axis for marker size
plt.scatter(x_axis, y_axis, s=[i*20 for i in y_axis], color="red", label='chicago')
# invert axis values
plt.gca().invert_yaxis() 
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

## Annotation enhancements

```python
# the following annotations
#Change the color of the markers to sky blue.
#Change the size of the markers to 5 times each data point.
#Add a legend for the city of Boston.
#Switch the axis so that the Fare($) data is on the x-axis---use y_,x_ insted of x_, y_
#Add a black edge color to the circles==edgecolors = "black"
#Make the linewidth of the circles 2 points.----linewidths = 2
plt.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color="skyblue", edgecolors = "black", linewidths = 2,label='Boston')

#Make the color 20% transparent.

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
#Add a title.
plt.title("PyBer Fare by Month", fontsize=15)
#Add a limit for the x-axis data ----this done after switching axis
plt.xlim(0, 45)
#Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis() 
```

```python
# the following annotations
#Change the color of the markers to sky blue.
#Change the size of the markers to 5 times each data point.
#Add a legend for the city of Boston.
#Switch the axis so that the Fare($) data is on the x-axis---use y_,x_ insted of x_, y_
#Add a black edge color to the circles==edgecolors = "black"
#Make the linewidth of the circles 2 points.----linewidths = 2
plt.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], marker = "d", color="skyblue", edgecolors = "black", linewidths = 2,label='Boston')

#Make the color 20% transparent.

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
#Add a title.
plt.title("PyBer Fare by Month", fontsize=15)
#Add a limit for the x-axis data ----this done after switching axis
plt.xlim(0, 45)
#Invert the y-axis so that January is at the top.
plt.gca().invert_yaxis()
```

## Pie charts

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Further annotations
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, startangle = 90, shadow = True, autopct='%.2f%%')  

plt.legend(title="Months", loc = (1, 0.7))

plt.show()
```

## Using the line chart

```python
%matplotlib inline
# import dependencies including STATISTICS
import matplotlib.pyplot as plt
import statistics
```

```python
# get stdev
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Create the plot.
# add the stdev and errorbar given by called yerr=stdev and add capsize to errorbar to cap them
plt.errorbar(x_axis, y_axis, marker="o", color="g", linewidth=2, yerr=stdev, capsize = 3, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```
