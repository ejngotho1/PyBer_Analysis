```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# object oriented
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# using show
plt.plot(x_axis, y_axis)
plt.title("PyBer Fare by Month")
plt.xlabel("Dates")
plt.ylabel("Fare($)")

# create label forlegend
plt.plot(x_axis, y_axis, Label = 'Boston')
# add legend
plt.legend()

plt.show()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="Red", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# lets do sme changes to above
# Create the plot.
plt.plot(x_axis, y_axis, marker=".", color="Green", linewidth=1, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# lets plot a bar chart
plt.bar(x_axis, y_axis)
```

```python
# lets annotate the graph
# creating a plot
plt.bar(x_axis, y_axis, color = 'green', label = 'Boston')

# labels for x and y
plt.xlabel("Date")
plt.ylabel('Fare($)')
plt.title("PyBer Fare y Month")
plt.legend()
```

```python
# creating a horizontal bar---plt.barh()
# creating a plot
plt.barh(x_axis, y_axis, color = 'green', label = 'Boston')

# labels for x and y
plt.ylabel("Date")
plt.xlabel('Fare($)')
plt.title("PyBer Fare y Month")
plt.legend()
```

```python
# lets scatter plot
# using MATLAB APPROACH
# using the ride data
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# import the matlab
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# using plt.plot()
plt.plot(x_axis, y_axis, 'o')
```

```python
# use scatter method
plt.scatter(x_axis, y_axis)
```

```python
# lets annotate
plt.scatter(y_axis, x_axis, color='red',label='Chicago')
plt.ylabel("Date")
plt.xlabel('Fare($)')
plt.title("PyBer Fare y Month")
plt.legend()
```

```python
plt.scatter(x_axis, y_axis)
```

```python
#  we want to make markers larger-- lets multiply by 3
# we will create an empty var and iterate thru the y-axis data
y_axis_larger = []

for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# lets plot that
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# lets use list comprehension which is faster and spots trend
new_list = plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# now we will use object oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# creating a burble
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=[i*5 for i in y_axis],  color='skyblue',edgecolors = 'black', linewidths=2, label='Boston')
#ax.title('PyBer Fare by Month')
ax.xlabel("Month")
ax.ylabel('Fare($)')
ax.lengend()
```

```python
# lets create pie charts
# autopct = "%1.1f%%"
#explode=[0,.1,0,0,0,0]
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# this will explode the 7th percntge
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%1.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
